```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import calibration_tools as tools
```

## Viewing how the free energies vary between repeats
### TIP3P

```python
plt.figure(figsize=[8,6])
for r in [1, 2, 3]:
    t3p = tools.AutoAnalyzeCalibration(['../sams/tip3p/out{0}.nc'.format(r)])
    max_salt = int(t3p.nsalt.max())
    plt.plot(-np.diff(t3p.sams[0]), label='SAMS, run {0}'.format(r), lw=2)
    plt.errorbar(x=range(max_salt), y=t3p.relative_free_energy, yerr=t3p.error_free_energy*2, fmt='o', label='BAR, run {0}'.format(r))

plt.xticks(range(0,20,1))
plt.title('TIP3P relative salt insertion free energies', fontsize=15)
plt.xlabel('Number of salt present', fontsize=16)
plt.ylabel('Relative free energy to add salt', fontsize=16)
plt.legend(fontsize=15)
plt.show()
```
